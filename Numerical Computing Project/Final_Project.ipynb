# Point of Intersection through <b><u><i><font color='red'>Newton Raphson</font></i></u></b>:

```python
def f(x):
    return 91.00425*x**2-1666050*x+7500426025 ## Assumed asteroid orbit
    
def Newton_Raphson(x):
    tol = 0.01
    count = 0
    h = f(x) / derivative(f,x) 

    while abs(h) >= tol: 
        h = f(x) / derivative(f,x)
        x = x - h
        count += 1
    print('The value of root converges at ' + str(count) + ' iterations')
    return x
    
#val <= 9153 --------> x = 7982 (-y)
#val > 9153 --------> x = 10324 (+y)
ini_value = 9154
x = Newton_Raphson(ini_value)
print('x = ', x)

y = 1.7321*x - 17321
print('y = ', y)
```
